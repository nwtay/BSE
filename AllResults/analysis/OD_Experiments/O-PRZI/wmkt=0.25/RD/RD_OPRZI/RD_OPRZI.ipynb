Results for RA model with O-PRZI traders

# Experiment 1

Parameters = (0.1, 0.2, 2, 0, 0.1)

## Opinion Spread

```python
from mpl_toolkits.mplot3d import Axes3D 
import matplotlib
import matplotlib.pyplot as plt
from matplotlib import cm, colors
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import numpy as np
import csv
from collections import defaultdict, Counter
from matplotlib.pyplot import *
import matplotlib.colors
from matplotlib.animation import FuncAnimation
```

```python
step = []
opinions = []
cs = []

# model_name = 'BC'

with open('opinionsRD_OPRZI_1.csv') as csv_file:
    readCSV = csv.reader(csv_file, delimiter=',')
    trial_count = 0
    time_period = 225
    for row in readCSV:
        trial_count = int(row[0][5:])
        ntime=row[1]
        for opinion in row[4:]:
            if opinion != ' ':
                opinions.append(float(opinion))
                step.append((trial_count-1)* time_period + float(ntime))
                
#     previous_time = 0
#     trial_count = 0
#     for row in readCSV:
#         if float(row[0]) >= previous_time:
#             transactions.append([((trial_count * time_period) + float(row[0])), float(row[1])])
#         else:
#             trial_count += 1
#         previous_time = float(row[0])
                
fig = plt.figure(3, figsize=(10,10))
plt.xlabel('Time', fontsize=30, labelpad=50)
plt.ylabel('Opinion', fontsize=30, labelpad=30)
# plt.title(model_name +" w=0.12", fontsize=30)
# p3 = scatter(step, smiddle, step, slower, step, supper)
ax=plt.axes()
ax.set_facecolor('black')
p3 = plt.scatter(step, opinions, color='orange')
# plt.xlim(xmin=0, xmax=180)
plt.ylim(ymin=-1, ymax=1)
plt.yticks(fontsize=24)
plt.xticks(fontsize=24, rotation=270)

plt.savefig('Graphs/opinionsRD_OPRZI_1.png',dpi=200,bbox_inches='tight',pad_inches=0)
plt.savefig('Graphs/opinionsRD_OPRZI_1.pdf',dpi=200,bbox_inches='tight',pad_inches=0)


plt.show()
```

## Transaction Spread

```python
n_trials = 15
time_period = 225
transactions = []

with open('transactionsRD_OPRZI_1.csv') as csv_file:
    readCSV = csv.reader(csv_file, delimiter=',')
    previous_time = 0
    trial_count = 0
    for row in readCSV:
        if float(row[0]) >= previous_time:
            transactions.append([((trial_count * time_period) + float(row[0])), float(row[1])])
        else:
            trial_count += 1
        previous_time = float(row[0])

time = np.array(transactions)[:,0]
price = np.array(transactions)[:,1]
  

# nums = np.arange(0, 600, 100)

print("Number of transactions:",len(price))

plt.figure(1, figsize=(10,10))
xlabel('Time', fontsize=30, labelpad=50)
ylabel('Price', fontsize=30, labelpad=30)
title("Transactions", fontsize=30)
# p1 = scatter(time, price, marker='x', color='red')
p1 = plot(time, price, 'orange', linewidth=3)
# p1 = plot(trials[i], 'green', linewidth=3)
p1 = plot()
ax = axes()
ax.set_facecolor('black')
# ylim(ymin=min(price)-20, ymax=max(price)+20)
ylim(ymin=0, ymax=200)
xlim(xmin=0, xmax=max(time)+20)
yticks(fontsize=30)
xticks(fontsize=30, rotation=270)
grid(linestyle='-', linewidth=1)

plt.savefig('Graphs/transactionsRD_OPRZI_1.png',dpi=200,bbox_inches='tight',pad_inches=0)
plt.savefig('Graphs/transactionsRD_OPRZI_1.pdf',dpi=200,bbox_inches='tight',pad_inches=0)


show()
```
