```python
n_trials = 15
time_period = 225
transactions = []

with open('transactionsBC_OPRZI_1.csv') as csv_file:
    readCSV = csv.reader(csv_file, delimiter=',')
    previous_time = 0
    trial_count = 0
    for row in readCSV:
        if float(row[0]) >= previous_time:
            transactions.append([((trial_count * time_period) + float(row[0])), float(row[1])])
        else:
            trial_count += 1
        previous_time = float(row[0])

time = np.array(transactions)[:,0]
price = np.array(transactions)[:,1]
  

# nums = np.arange(0, 600, 100)

print("Number of transactions:",len(price))

plt.figure(1, figsize=(10,10))
xlabel('Time', fontsize=30, labelpad=50)
ylabel('Price', fontsize=30, labelpad=30)
title("Transactions", fontsize=30)
# p1 = scatter(time, price, marker='x', color='red')
p1 = plot(time, price, 'orange', linewidth=3)
# p1 = plot(trials[i], 'green', linewidth=3)
p1 = plot()
ax = axes()
ax.set_facecolor('black')
# ylim(ymin=min(price)-20, ymax=max(price)+20)
ylim(ymin=0, ymax=200)
xlim(xmin=0, xmax=max(time)+20)
yticks(fontsize=30)
xticks(fontsize=30, rotation=270)
grid(linestyle='-', linewidth=1)
show()
```
